### Importing Libraries

import seaborn as sns
import matplotlib.pyplot as plt

### Load Dataset

```python
fmri = sns.load_dataset('fmri')
fmri
```

### Regression plot (scatter)

```python
sns.relplot(x='timepoint',y='signal',hue='event',data=fmri,aspect=3)
```

```python
sns.relplot(x='timepoint',y='signal',hue='region',data=fmri,aspect=3)
```

### Line plot

```python
sns.relplot(x='timepoint',y='signal',hue='event',data=fmri,aspect=3,kind='line')
```

```python
sns.relplot(x='timepoint',y='signal',hue='region',data=fmri,aspect=3,kind='line')
```

### Box plot

```python
sns.boxplot(x='timepoint',y='signal',hue='region',data=fmri)
```

```python
sns.boxplot(x='timepoint',y='signal',hue='event',data=fmri)
```

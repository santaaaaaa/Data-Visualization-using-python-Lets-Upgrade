### import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

###  Question 1
#### Plot a line plot between a and b: 

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
plt.plot(a,b,'g--*')
plt.xlabel('Value range from 40 to 50')
plt.ylabel('Value range from 50 to 60')
plt.title('Line Plot')
```

### Question 2
#### Plot a line plot showing the sales trend in company 1 and 2:

```python
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2
```

```python
plt.plot(days,sales_1,'g', label = 'Company 1')
plt.plot(days,sales_2,'b',label = 'Company 2')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.title('Sales of two companies in a week')
plt.legend(loc='lower right')
```

### Question 3
#### Create 3 by 3 subplots:

```python
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]
```

```python
plt.subplot(3,3,1)
plt.plot(x,y1)
plt.title('Y1')
plt.subplot(3,3,2)
plt.plot(x,y2)
plt.title('Y2')
plt.subplot(3,3,3)
plt.plot(x,y3)
plt.title('Y3')
plt.subplot(3,3,4)
plt.plot(x,y4)
plt.title('Y4')
plt.subplot(3,3,5)
plt.plot(x,y5)
plt.title('Y5')
```
